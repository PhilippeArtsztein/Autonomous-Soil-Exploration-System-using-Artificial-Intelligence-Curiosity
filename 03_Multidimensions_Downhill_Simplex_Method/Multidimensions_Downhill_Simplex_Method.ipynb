The Nelder–Mead method was proposed by John Nelder and Roger Mead in 1965.<br>
The method is also known as the Downhill Simplex Method.<br>
Downhill Simplex is a numerical method used to find the minimum or maximum of an<br>
objective function in a multidimensional space.<br>
It is a direct search method (based on function comparison) and is often applied<br>
to nonlinear optimization problems for which derivatives may not be known.<br>
The method may frequently be the best method to use if the figure of merit is<br>
“get something working quickly” for a problem whose computational burden is small.<br>
The method has a geometrical naturalness about it which makes it delightful to describe<br>
or work through.</div>

<div class="alert alert-success">
<b><u>DownhillSimplexMethodInMultidimensions</u></b><br>
Is a implementation of Downhill Simplex Method fuction to minimize our objective fanction.<br><br><b><u>Input</u></b><br>
<u><i>ObjectiveFunction</i></u><br>
&emsp;Is the name of our objective function we wish to minimize. The objective function<br>
&emsp;must return value (a score).<br>
<u><i>XX_Initial</i></u><br>
&emsp;The XX_Initial array are the variables of the function. The array contain an initial<br>
&emsp;guess, whitch is the starting point. The size of the array must be equal to the number<br>
&emsp;of variables.<br>
<u><i>Parameters</i></u><br>
&emsp;The Parameters array stores the fix constants of the objective function. The size of<br>
&emsp;the array must be equal to the number of parameters.<br>
<u><i>Print_Enable</i></u><br>
&emsp;If sets to <b>True</b> the function will print intermediate variables and scores of convergence<br>
&emsp;to the result. If sets to <b>False</b> the function will be in silent mode<br>
<u><i>Initial_Step</i></u> (initially set to 0.1)<br>
&emsp;To calculate additional vertex points to form a nth dimension vertex corners, we use the<br> 
&emsp;XX_Initial, and add nodes vertex points distence of <i>Initial_Step</i> from <i>XX_Initial</i>.<br> 
&emsp;<b>Note :</b>This value corresponds to each objective function in a different way, there is  <b>no</b><br>
&emsp;rule of thumb to estimate its value, so you have to experiment until you get the optimal<br> 
&emsp;value for any given problem.<br>
<u><i>Non_Ameliorated_Max</i></u> (initially set to 10), <i>Non_Ameliorated_Threshold</i> (initially set to 10e-6)<br>
&emsp;Return the best results so far after <i>Non_Ameliorated_Max</i> iteration when<br> &emsp;thers no score amelioration in context of <i>Non_Ameliorated_Threshold</i> value<br>
<u><i>Max_Iteration_Allowed</i></u> (initially set to 500)<br>
&emsp;The maximum number of iterations until the search stops.<br>
<br><b><u>Output</u></b><br>
&emsp;The function returns the best values of variables array which gives the best result (score).<br>
</div>

```python
#====================================================================================
def DownhillSimplexMethodInMultidimensions(ObjectiveFunction, 
                                           XX_Initial, 
                                           Parameters,
                                           Print_Enable, 
                                           Initial_Step=0.1,
                                           Non_Ameliorated_Max=10, 
                                           Non_Ameliorated_Threshold=10e-6, 
                                           Max_Iteration_Allowed=500):
    # Initial Settings of The values of the coefficients for 
    # Reflection, Contraction, Shrinkage of the multidimensional vertex corners.
    Alpha                = 1.0
    Gamma                = 2.0 
    Rho                  =-0.5 
    Sigma                = 0.5
    # Make a referance vertex node
    XX_Dinamtion         = len(XX_Initial)
    Old_Best             = ObjectiveFunction(XX_Initial, Parameters)
    Non_Ameliorated_Index= 0
    Results              = [[XX_Initial, Old_Best]]
    # Calculate all additional nth dimension vertex points and scores
    for iii in range(XX_Dinamtion):
        XX     = copy.copy(XX_Initial)
        XX[iii]= XX[iii] + Initial_Step
        Score  = ObjectiveFunction(XX, Parameters)
        Results.append([XX, Score])
        
    # Simplex Iteration Loop
    Iterations = 0
    if(Print_Enable==True):
        print("\033[4mBest so far:\033[0m")
    while 1:
        ##== First Action: Order according to the values at the vertices (Scors)
        Results.sort(key=lambda XX: XX[1])
        New_Best= Results[0][1]

        #== Now, Maybe this is the place to return results 
        # Break and Return Results after Max_Iteration_Allowed
        if(Iterations >= Max_Iteration_Allowed):
            return(False, Results[0])
        Iterations += 1
        # Store the last improvement
        if(Print_Enable==True):
            print("x= {: .4E}".format(Results[0][0][0]),
                  "  y= {: .4E}".format(Results[0][0][1]),
                  "  Score= {: .4E}".format(New_Best))
        if(New_Best < Old_Best-Non_Ameliorated_Threshold):
            Non_Ameliorated_Index = 0
            Old_Best              = New_Best
        else:
            Non_Ameliorated_Index+= 1
        # Break after Non_Ameliorated_Max iterations with no improvement
        if(Non_Ameliorated_Index >= Non_Ameliorated_Max):
            return(True, Results[0])

        ##== Second Action: Calculate Centroid (geometric center)
        XX_0= [0.0] * XX_Dinamtion
        for Tup in Results[:-1]:
            for iii, CC in enumerate(Tup[0]):
                XX_0[iii] += CC / (len(Results)-1)

        #== First possible reshaping: Reflection, Compute reflected point
        #                             and get Rscore a reflection away from
        #                             the high point
        XR= [0.0] * XX_Dinamtion
        for iii in range(XX_Dinamtion):
            XR[iii] = XX_0[iii] + Alpha*(XX_0[iii] - Results[-1][0][iii])
        Rscore= ObjectiveFunction(XR, Parameters)
        if(Results[0][1] <= Rscore < Results[-2][1]):
            del Results[-1]
            Results.append([XR, Rscore])
            continue

        #== Second possible reshaping: Expansion or Reflection and get
        #                              Escore or Rscore A reflection 
        #                              and expansion away from the high point
        if(Rscore < Results[0][1]):
            XE= [0.0] * XX_Dinamtion
            for iii in range(XX_Dinamtion):
                XE[iii] = XX_0[iii] + Gamma*(XX_0[iii] - Results[-1][0][iii])
            Escore = ObjectiveFunction(XE, Parameters)
            if(Escore < Rscore):
                del Results[-1]
                Results.append([XE, Escore])
                continue
            else:
                del Results[-1]
                Results.append([XR, Rscore])
                continue

        #== Third possible reshaping: Contraction and get Cscore
        #                             A contraction along one dimension
        #                             from the high point
        XC= [0.0] * XX_Dinamtion
        for iii in range(XX_Dinamtion):
            XC[iii] = XX_0[iii] + Rho*(XX_0[iii] - Results[-1][0][iii])
        Cscore = ObjectiveFunction(XC, Parameters)
        if(Cscore < Results[-1][1]):
            del Results[-1]
            Results.append([XC, Cscore])
            continue

        #== Fourth possible reshaping: Shrink (multiple contraction)
        #                              A contraction along all dimensions 
        #                              towards the low point
        XX_1= Results[0][0]
        XS= []
        for Tup in Results:
            XT= [0.0] * XX_Dinamtion
            for iii in range(XX_Dinamtion):
                XT[iii]= XX_0[iii] + Sigma*(Tup[0][iii] - XX_1[iii])
            Score= ObjectiveFunction(XT, Parameters)
            XS.append([XT, Score])
        Results= XS
```

<div class="alert alert-danger">
To test the code and show its usage, I am using <u>Rosenbrock function</u>.<br>
In mathematical optimization, the Rosenbrock function is a non-convex function,<br> 
introduced by Howard H. Rosenbrock in 1960, which is used as a performance test<br>
problem for optimization algorithms. It is also known as Rosenbrock's valley or<br>
Rosenbrock's banana function.<br>
The global minimum is inside a long, narrow, parabolic shaped flat valley.<br>
To find the valley is trivial. To converge to the global minimum, however, is difficult.<br>
The function is defined by $f\left(x,y\right)=\left(a-x\right)^{2}+b\left(y-x^{2}\right)^{2}$<br>

<div class="alert alert-success">
<b><u>ObjectiveFunction</u></b><br>
Is the function we are defining and we are interested in finding its minimum.<br> 
In case when the parameters are fixed and defined as $a=1$ and $b=100$,<br>
The minimum should be at $x=1$ and $y=1$ and the value of the function should be 0<br>
<br><b><u>Input</u></b><br>
<u><i>XX</i></u><br>
&emsp;The XX array are the variables of the function. The array contain an<br>
&emsp;initial guess, whitch is the starting point. The size of the array<br>
&emsp;is equal to the number of variables.<br>
<u><i>Parameters</i></u><br>
&emsp;The Parameters array are fixed constants.<br>
&emsp;The size of the array is equal to the number of parameters.<br>
<br><b><u>Output</u></b><br>
&emsp;The function returns the value of the function given the values of the variables<br>
&emsp;&emsp;<u><i>Result[0]</i></u><br>
&emsp;&emsp;&emsp;<b>True</b> if the algorithm successfully converged to a minimum<br>
&emsp;&emsp;&emsp;<b>False</b> if the algorithm failed to converge to any minimum<br>
&emsp;&emsp;<u><i>OptimizedResult[1][0][0]</i></u>....<u><i>OptimizedResult[1][0][n]</i></u><br>
&emsp;&emsp;&emsp;All n variables values where the objective function is minimum.<br> 
&emsp;&emsp;<u><i>OptimizedResult[1][1]</i></u><br>
&emsp;&emsp;&emsp;The value of the objective function to its minimum.
</div>

```python
def ObjectiveFunction(XX, Parameters):
    return((Parameters[0]-XX[0])**2 + Parameters[1]*(XX[1]-(XX[0])**2)**2)
```

<br><b><u>Input</u></b><br>
Variables of the function with initial guess<br>
&emsp;<i>ObjectiveFunction</i> = The objective function name<br>
&emsp;<i>XX</i> = [-2.0, -1.0]<br>
&emsp;<i>PP</i> = [1.0,100.0]<br>

```python
PP= [1.0,100.0]
#XX= [3.0, 2.0]
#XX= [-20.0, -10.0]
# Lets guess a starting point (or Initial Guess) x=-2 and y=-1
XX= [-2.0, -1.0]

OptimizedResult= DownhillSimplexMethodInMultidimensions(ObjectiveFunction= 
                                                        ObjectiveFunction, 
                                                        XX_Initial= XX, 
                                                        Parameters= PP, 
                                                        Print_Enable= True)
print("\n\033[4mFinal results:\033[0m")
if(OptimizedResult[0]==True):
    print("The algorithm successfully converged to a minimum,")
    print("The best results obtained is:")
    print("x= {: .4E}".format(OptimizedResult[1][0][0]),
          "  y= {: .4E}".format(OptimizedResult[1][0][1]),
          "  Score= {: .4E}".format(OptimizedResult[1][1]))
elif(OptimizedResult[0]==False):
    print("The algorithm failed to converge to any minimum,")
    print("even after 500 Iteration,")
    print("The last results it reached is:")
    print("x= {: .4E}".format(OptimizedResult[1][0][0]),
          "  y= {: .4E}".format(OptimizedResult[1][0][1]),
          "  Score= {: .4E}".format(OptimizedResult[1][1]))
```
